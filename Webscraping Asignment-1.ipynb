# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&view=simple') 
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
page1=requests.get('https://www.imdb.com/search/title/?groups=top_100&view=simple&sort=user_rating,desc&start=51&ref_=adv_nxt') 
page1
```

```python
soup1=BeautifulSoup(page1.content)
```

```python
movie_titles=[]     

for i in soup.find_all('span',class_="lister-item-header"):
    movie_titles.append(i.text[5:-9])
for i in soup1.find_all('span',class_="lister-item-header"):
    movie_titles.append(i.text[5:-9])
```

```python
movie_titles
```

```python
ratings=[]     

for i in soup.find_all('div',class_="col-imdb-rating"):
    ratings.append(i.text[30:-26])

for i in soup1.find_all('div',class_="col-imdb-rating"):
    ratings.append(i.text[30:-26])
```

```python
ratings
```

```python
YOR=[]
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    YOR.append(i.text[1:5])

for i in soup1.find_all('span',class_="lister-item-year text-muted unbold"):
    YOR.append(i.text[1:5])
```

```python
YOR
```

```python
import pandas as pd
TOP100=pd.DataFrame({'Titles':movie_titles,'Ratings':ratings,'Release Year':YOR})
```

```python
TOP100
```
